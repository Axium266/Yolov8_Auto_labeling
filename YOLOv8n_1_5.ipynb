```python
#@title 模型訓練 ( 繞過 Torch 限制 )
model = "/content/yolov8.yaml" #@param {type:"string"}
data = "/content/data.yaml" #@param {type:"string"}

# 關閉 Torch 安全性限制
import torch
from ultralytics import YOLO

_real_torch_load = torch.load

def patched_load(*args, **kwargs):
    kwargs['weights_only'] = False
    return _real_torch_load(*args, **kwargs)

torch.load = patched_load

# 載入 Yolov8 模型
model = YOLO(model)

# 開始訓練
model.train(data=data, epochs=200, batch=16, imgsz=640, workers=4)
```

```python
#@title 測試模型
best = "/content/best.pt" #@param {type:"string"}
image = "/content/1.png" #@param {type:"string"}

# 偵測圖片
from ultralytics import YOLO
model = YOLO(best)
results = model(image, conf=0.5)

# 顯示辨識結果
for result in results:
    result.show()
```
